```python
# Imports (data handling + saving)
import pandas as pd
import numpy as np
from pathlib import Path
import pickle

# Paths
DATA_RAW = Path("../data/raw")
DATA_PROCESSED = Path("../data/processed")

# Create processed directory if it doesn't exist
DATA_PROCESSED.mkdir(parents=True, exist_ok=True)
```

```python
#Load raw datasets
train_df = pd.read_csv(DATA_RAW / "train.csv")
train_demo_df = pd.read_csv(DATA_RAW / "train_demographics.csv")
test_df = pd.read_csv(DATA_RAW / "test.csv")
test_demo_df = pd.read_csv(DATA_RAW / "test_demographics.csv")
```

```python
# Basic Sanity Checks
print("Train shape:", train_df.shape)
print("Train demographics shape:", train_demo_df.shape)
print("Test shape:", test_df.shape)
print("Test demographics shape:", test_demo_df.shape)
```

```python
train_df.info()
train_demo_df.info()
```

```python
# Save ingested raw datasets separately for downstream notebooks

with open(DATA_PROCESSED / "train.pkl", "wb") as f:
    pickle.dump(train_df, f)

with open(DATA_PROCESSED / "train_demographics.pkl", "wb") as f:
    pickle.dump(train_demo_df, f)

with open(DATA_PROCESSED / "test.pkl", "wb") as f:
    pickle.dump(test_df, f)

with open(DATA_PROCESSED / "test_demographics.pkl", "wb") as f:
    pickle.dump(test_demo_df, f)

print("Raw datasets saved individually to data/processed/")
```
