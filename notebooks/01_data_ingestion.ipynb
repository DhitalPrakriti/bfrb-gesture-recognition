## 1. Imports and Path Configuration

This section imports required libraries and defined the standerized paths for raw input dta and processed artifacts used throughout the pipeline.

```python
# Imports (data handling + saving)
import pandas as pd
import numpy as np
from pathlib import Path
import pickle

# Paths
DATA_RAW = Path("../data/raw")
DATA_PROCESSED = Path("../data/processed")

# Create processed directory if it doesn't exist
DATA_PROCESSED.mkdir(parents=True, exist_ok=True)
```

## 2. Load Raw datasets

In this step, all CSV datasets are loaded from the `data/raw` directory without any modification. These datasets serve as the immutable input for downstream analysis and preprocessing.

```python
#Load raw datasets
train_df = pd.read_csv(DATA_RAW / "train.csv")
train_demo_df = pd.read_csv(DATA_RAW / "train_demographics.csv")
test_df = pd.read_csv(DATA_RAW / "test.csv")
test_demo_df = pd.read_csv(DATA_RAW / "test_demographics.csv")
```

## 3. Data Integrity Checks

Basic sanity checks are performed to verify dataset dimensions and ensure that the files were loaded correctly. No transformations are applied at this stage.

```python
# Basic Sanity Checks
print("Train shape:", train_df.shape)
print("Train demographics shape:", train_demo_df.shape)
print("Test shape:", test_df.shape)
print("Test demographics shape:", test_demo_df.shape)
```

## 4. Schema Inspection

This step inspects column names, data types, and memory usage to document the structure of the raw datasets. This information supports reproducibility and debugging in later stages.

```python
train_df.info()
train_demo_df.info()
```

## 5. Processed Artifacts

The raw datasets are serialized and saved individually to the `data/processed` directory. These artifacts act as standardized inputs for EDA, preprocessing, and modeling notebooks.

```python
# Save ingested raw datasets separately for downstream notebooks

with open(DATA_PROCESSED / "train.pkl", "wb") as f:
    pickle.dump(train_df, f)

with open(DATA_PROCESSED / "train_demographics.pkl", "wb") as f:
    pickle.dump(train_demo_df, f)

with open(DATA_PROCESSED / "test.pkl", "wb") as f:
    pickle.dump(test_df, f)

with open(DATA_PROCESSED / "test_demographics.pkl", "wb") as f:
    pickle.dump(test_demo_df, f)

print("Raw datasets saved individually to data/processed/")
```
